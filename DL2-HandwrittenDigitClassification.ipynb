```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import image
import tensorflow as tf
from tensorflow import keras
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

```python
#lets see one of the record
x_train[0]
```

```python
plt.imshow(x_train[0])
plt.show()
print(y_train[0])
```

```python
#lets standardize the data and bring the value range between 0 to 1, so it becomes easy for the model to perform calculation on smaller data
x_train = x_train/255
x_test = x_test/255
```

```python
x_train[0]
```

```python
#now lets see the frequency of each digit in the record
unique_values, counts = np.unique(y_train, return_counts=True)
print("Unique values:", unique_values)
print("Counts:", counts)
```

```python
#there is an equal distribution of the data, and the data is also now standrdized
#data is also in grayscale format, and also no need for resizing, lets train the model
model = keras.Sequential(layers=[
    keras.layers.Flatten(input_shape=((28,28))),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
                                 ])
```

```python
#lets compile it with all hyper paramters
#adam optimizer updates the weight during trainig
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#lets fit the data in the model
model.fit(x_train, y_train, epochs=10)
```

```python
#so on training data, it has 98% accuracy, lets see the same on test data
loss, accuracy = model.evaluate(x_test, y_test)
print(loss, accuracy)
```

```python
#so the prediction on the test data is : 96%
#lets predict on the test data, and see the results
y_test_pred = model.predict(x_test)
y_test_pred.shape
```

```python
#we have 10 columnd, cause for each unique value i.e 0-9 in label, it has given us the probaility value for each,
#which so ever is max, we will choose that.
y_test_pred_result = [np.argmax(i) for i in y_test_pred]
print(y_test_pred_result[:10])
```

```python
print(y_test[:10])
```

```python
#lets plot the actual value against predicted value on confusion matrix, and see the result,we are getting.
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_test_pred_result)
cm
```

```python
import seaborn as sns
sns.set()
plt.figure(figsize=(10, 7)) # Increase figure size
sns.heatmap(cm, annot=True, fmt='d', cmap='viridis') # Change colormap to viridis and add title
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
